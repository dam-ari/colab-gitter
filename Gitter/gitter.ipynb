## **imports**: Run on load

#### imports 

```python
import subprocess as p
import signal
import time
from threading import Thread
from google.colab import output
```

#### getInput 

```python
def getInput(inputMsg,timeout=15):
  var = None

  def check():
      time.sleep(timeout)
      if var != None:
          return var
      output.eval_js('new AbortExecution();') # Stop the execution of cells
      

  Thread(target = check).start()

  var = input(inputMsg)
  if (str(var).find('\'error\': \'ReferenceError: AbortExecution is not defined\'')>0):
    var = None
  return var
```
